```python
import bs4 as bs
import pickle 
import requests 
import lxml 

def tickersFTSE(): 
    resp = requests.get('https://en.wikipedia.org/wiki/FTSE_100_Index#Constituents_in_March_2020')
    soup = bs.BeautifulSoup(resp.text)
    table = soup.findAll('table', {'class': 'wikitable sortable'})[1]
    tickers = []
    for row in table.findAll('tr')[1:]: 
        ticker = row.findAll('td')[1].text
        ticker = ticker.replace("\n","")
        tickers.append(ticker)
        
    with open('ftse100tickers.pickle', 'wb') as f: 
        pickle.dump(tickers, f)
    
    return tickers 

tickersFTSE()
```

```python
ftse100 = pd.DataFrame(tickersFTSE())
ftse100.to_csv('ftse100.csv', index=False, header=False)
pd.read_csv('ftse100.csv', header=None)
```

```python
def opInfo(ticker='BARC'):
    return ek.get_data('0#'+ticker+'*.EX',['STRIKE_PRC','EXPIR_DATE','PUTCALLIND','CF_CLOSE'])[0]

def getOpPrices(ticker='BARC'):
    fst = True 
    instr = list(set(ek.get_data('0#'+ticker+'*.EX',['STRIKE_PRC','EXPIR_DATE','PUTCALLIND','CF_CLOSE'])[0]['Instrument']))
    for i in instr: 
        try: 
            if fst:
                a = ek.get_timeseries(i, 'CLOSE',
                                  start_date='20100101',end_date='20200601').reset_index()
                a.insert(0, 'OPTION_ID', a.columns.name)
                fst = False
            else: 
                b = ek.get_timeseries(i,'CLOSE',
                                  start_date='20100101',end_date='20200601').reset_index()
                b.insert(0, 'OPTION_ID', b.columns.name)
                a = a.append(b)
        except: 
            pass
        time.sleep(0.2)
    return a
```

```python
fst = True 
for t in stcks:
    if fst: 
        sdData = pd.DataFrame(ek.get_timeseries(t, start_date='2013-12-01', end_date='2018-12-01')['CLOSE'])
        sdData['Ticker'] = t.replace('.L','')
        fst = False
    else: 
        d = pd.DataFrame(ek.get_timeseries(t, start_date='2013-12-01', end_date='2018-12-01')['CLOSE'])
        d['Ticker'] = t.replace('.L','')
        sdData = sdData.append(d)
        time.sleep(0.2)

sdData.columns = ['Price', 'Ticker']
sdData['SRet'] = np.log(sdData.Price / sdData.Price.shift())
sdData

sdrets = sdData.reset_index().pivot('Date', 'Ticker', 'SRet')
sdrets

sd = pd.DataFrame(sdrets.std()).reset_index()
sd['Mean'] = pd.DataFrame(sdrets.mean()).reset_index()[0]
sd.columns = ['Ticker', 'SD', 'SMean']
sd

pd.merge(sData.reset_index(), sd)
```
